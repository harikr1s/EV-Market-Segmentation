```python
X = df.iloc[:, [5, 13]].values
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42, n_init = 'auto')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42, n_init = 'auto')
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, label = 'Centroids')
plt.title('Clusters of EV')
plt.xlabel('Efficieny (Wh/km)')
plt.ylabel('PriceINR (in thousands)')
plt.legend()
plt.savefig('kmeans.png')
```

We can see that lower priced EVs have higher efficiency
